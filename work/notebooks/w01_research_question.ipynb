## 2. The question: decision, action, cost of a wrong call

**Research question:** Which measurable signals are most useful for identifying pages at risk of decline so review effort can be prioritized effectively?

**Unit of analysis:** One content page (`content_id`) at the current snapshot.

**Output:** A ranked signal analysis summary with evidence-backed priority signals for page review.

**Who acts and what action:** A content editor or SEO analyst reviews high-priority pages first and chooses actions such as refresh, rewrite, improve structure, or monitor.

**Cost of a wrong recommendation:**
- False positive: editor time is wasted on low-risk pages with limited upside.
- False negative: high-risk pages are missed, allowing further decline and opportunity loss.

## 3. Quick look at the data (2-3 real numbers)

The starter dataset supports this lane with practical scope:

- It contains **30,000 rows**, **44 columns**, and **32 unique clients**.
- A substantial share of pages are currently marked as declining (`trend_direction = "down"`), so risk-focused signal triage is relevant.
- `avg_position = 0` appears in many rows (no position data), which confirms that realistic signal analysis must include data-quality handling.

These observations indicate that Ranking Signal Analysis is a valid and useful lane for deeper work in the coming weeks.

```python
import pandas as pd

csv_path = '/home/runner/work/Fly-Rank-week1-ASSIGNMENT/Fly-Rank-week1-ASSIGNMENT/data/raw/content_refresh_anonymized.csv'
df = pd.read_csv(csv_path)

rows = len(df)
cols = df.shape[1]
clients = df['client_id'].nunique()
declining_rate = (df['trend_direction'] == 'down').mean() * 100
avg_pos_zero = (df['avg_position'] == 0).sum()

print(f'Rows: {rows:,}')
print(f'Columns: {cols}')
print(f'Unique clients: {clients}')
print(f'Declining pages (trend_direction=="down"): {declining_rate:.2f}%')
print(f'Rows with avg_position == 0 (no position data): {avg_pos_zero:,}')
```
