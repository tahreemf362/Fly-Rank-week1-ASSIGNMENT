## 2. The question: decision, action, cost of a wrong call

**Search question:** Which content pages should be prioritized for refresh in the next review cycle to reduce risk of further performance decline?

**Unit of analysis:** One content page (`content_id`) at the current snapshot.

**Output:** A ranked priority score / queue of pages to review first, with simple reason codes from observable signals.

**Decision improved:** Prioritization of editorial refresh effort.

**Who acts and what action:** Content editor or SEO analyst reviews top-ranked pages first and chooses action (refresh, rewrite sections, monitor, or defer).

**Cost of a wrong recommendation:**
- False positive: editor time is wasted on pages with low real upside.
- False negative: high-risk pages are missed and may continue declining, causing opportunity loss.

**Why data/ML can help:** The decision depends on multiple interacting signals (impressions, clicks, CTR, position, engagement, age, trend). A single hand-written rule is transparent and useful as a baseline, but it may miss nonlinear patterns and interactions that improve top-of-queue precision.
